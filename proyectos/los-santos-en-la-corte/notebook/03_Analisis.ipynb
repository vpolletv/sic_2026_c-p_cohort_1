Asignacion de Nombre a las regiones:

```python
regiones = {
    1: "Tarapacá",
    2: "Antofagasta",
    3: "Atacama",
    4: "Coquimbo",
    5: "Valparaíso",
    6: "O'Higgins",
    7: "Maule",
    8: "Biobío",
    9: "La Araucanía",
    10: "Los Lagos",
    11: "Aysén",
    12: "Magallanes",
    13: "Metropolitana",
    14: "Los Ríos",
    15: "Arica y Parinacota",
    16: "Ñuble"
}

df_new["region_nombre"] = df_new["region"].map(regiones)
```

Analaisis de Muestra por varible de genero:


```python
df["sexo"].value_counts(dropna=False)
```

```python
sexo_map = {
    1: "Hombre",
    2: "Mujer"
}

df["sexo_nombre"] = df["sexo"].map(sexo_map)
```

 Analisis de formalidad vs formalidad por genero:

Creacion de columna "situacion":

```python
df["situacion"] = df["informalidad"].map({0: "Formal", 1: "Informal"})
```

```python
tabla = pd.crosstab(df["sexo_nombre"], df["situacion"])
print(tabla)
```

```python
ax = tabla.plot(kind="bar", figsize=(6,4), color=["#1D9E75", "#D85A30"])
ax.set_title("Formales vs. Informales por sexo")
ax.set_xlabel("")
ax.set_ylabel("Cantidad de microemprendedores")
ax.legend(title="Situación")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
tabla_long = tabla.reset_index().melt(
    id_vars="sexo_nombre", var_name="situacion", value_name="cantidad"
)

fig = px.bar(
    tabla_long,
    x="sexo_nombre", y="cantidad", color="situacion",
    barmode="group", text="cantidad",
    color_discrete_map={"Formal": "#1D9E75", "Informal": "#D85A30"},
    labels={"sexo_nombre": "", "cantidad": "Cantidad de microemprendedores", "situacion": "Situación"},
    title="Formales vs. Informales por sexo"
)
fig.update_traces(textposition="outside")
fig.show()
```

```python
tabla_pct = pd.crosstab(df["sexo_nombre"], df["situacion"], normalize="index") * 100
tabla_pct = tabla_pct.round(1)
print(tabla_pct)
```

```python
tabla_long = tabla_pct.reset_index().melt(
    id_vars="sexo_nombre", var_name="situacion", value_name="porcentaje"
)

fig = px.bar(
    tabla_long,
    x="sexo_nombre", y="porcentaje", color="situacion",
    barmode="group", text="porcentaje",
    color_discrete_map={"Formal": "#1D9E75", "Informal": "#D85A30"},
    labels={"sexo_nombre": "", "porcentaje": "% del grupo", "situacion": "Situación"},
    title="Tasa de formalidad/informalidad por sexo"
)
fig.update_traces(texttemplate="%{text}%", textposition="outside")
fig.update_layout(yaxis_range=[0, 70])
fig.show()
```

Analisis de informalidad por region:


Conteo de formales vs informales por región

```python
tabla_n = pd.crosstab(df["region_nombre"], df["situacion"])
tabla_n["Total"] = tabla_n.sum(axis=1)
```

Porcentaje dentro de cada región (la comparación correcta, igual que con sexo):

```python
tabla_pct = (pd.crosstab(df["region_nombre"], df["situacion"], normalize="index") * 100).round(1)
```

Unir ambas en una sola tabla

```python
tabla = tabla_n.join(tabla_pct, rsuffix="_%")
tabla = tabla.sort_values("Informal_%", ascending=False)
print(tabla[["Formal", "Informal", "Total", "Formal_%", "Informal_%"]])
```

```python
grafico_pct = tabla.reset_index()

grafico_pct = grafico_pct.melt(
    id_vars=grafico_pct.columns[0],
    value_vars=["Formal_%", "Informal_%"],
    var_name="Estado",
    value_name="Porcentaje"
)

fig = px.bar(
    grafico_pct,
    x=grafico_pct.columns[0],
    y="Porcentaje",
    color="Estado",
    barmode="stack",
    text_auto=".1f",
    title="Porcentaje de formalidad por region"
)

fig.show()
```
