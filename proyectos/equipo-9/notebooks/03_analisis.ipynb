```python
# Análisis Principal
import pandas as pd

# 1. Cargar el dataset limpio
df = pd.read_csv('../data/dataset_limpio.csv')

# 2. Agrupación simple usando los nombres correctos
resumen = df.groupby('category')['current_price'].sum().reset_index()

print("Resumen de ingresos por categoría:")
display(resumen)

# Conclusión: Ciertas categorías generan más ingresos bajo estrategias de descuento.
```
