```python
import pandas as pd

# 1. Usamos los nombres exactos y en minúsculas de tu archivo
columnas_necesarias = ['category', 'current_price', 'markdown_percentage', 'size']

# 2. Cargar el dataset optimizado
df = pd.read_csv('../data/dataset_original.csv', usecols=columnas_necesarias)

# 3. Eliminar nulos
df.dropna(subset=['category', 'current_price'], inplace=True)

# 4. Eliminar duplicados
df.drop_duplicates(inplace=True)

# 5. Exportar el archivo
df.to_csv('../data/dataset_limpio.csv', index=False)
print("Datos limpios exportados de forma optimizada.")
```
